```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
```

```python
# Generate synthetic data
# Set n_informative=2 (all features are informative) and n_redundant=0, n_repeated=0
X, y = make_classification(
    n_samples=100,
    n_features=2,
    n_informative=2,  # All features are informative
    n_redundant=0,    # No redundant features
    n_repeated=0,     # No repeated features
    n_classes=2,
    n_clusters_per_class=1,
    random_state=42
)
```

```python
cov_matrix = np.cov(X.T)

eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

projected_data = X.dot(eigenvectors[:, 0])
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='viridis', label="Data Points")
plt.quiver(0, 0, eigenvectors[0, 0], eigenvectors[1, 0], angles='xy', scale_units='xy', scale=1, color='r', label="Eigenvector 1")
plt.quiver(0, 0, eigenvectors[0, 1], eigenvectors[1, 1], angles='xy', scale_units='xy', scale=1, color='b', label="Eigenvector 2")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.title("EigenClass: Data and Eigenvectors")
plt.show()
```
