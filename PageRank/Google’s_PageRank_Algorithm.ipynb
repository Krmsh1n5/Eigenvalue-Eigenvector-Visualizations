```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
G = nx.DiGraph()
```

```python
G.add_nodes_from(range(5))
```

```python
G.add_edges_from([(0, 1), (1, 2), (2, 0), (2, 3), (3, 4), (4, 3)])
```

```python
pagerank_scores = nx.pagerank(G)
```

```python
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightblue', node_size=2000, font_size=10, font_weight='bold', edge_color='gray')
plt.title("PageRank: Web Graph")
plt.show()
```

```python
plt.bar(pagerank_scores.keys(), pagerank_scores.values(), color='orange')
plt.xlabel("Node (Web Page)")
plt.ylabel("PageRank Score")
plt.title("PageRank Scores")
plt.show()
```
