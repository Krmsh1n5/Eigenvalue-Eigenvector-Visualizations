```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_swiss_roll
from sklearn.decomposition import KernelPCA
```

```python
# Generate the Swiss Roll Dataset
np.random.seed(42)
n_samples = 2000
X, t = make_swiss_roll(n_samples, noise=0.0)


fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(projection='3d')
scatter = ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=t, s=10, cmap='hot_r')  # Capture the scatter plot object
ax.set_xlabel('X', fontsize=12)
ax.set_ylabel('Y', fontsize=12)
ax.set_zlabel('Z', fontsize=12)
ax.view_init(elev=7, azim=-80)  # Adjust viewing angle
plt.title("Original Swiss Roll Dataset (3D)", fontsize=14)
plt.colorbar(scatter, label='t')  # Pass the scatter object to colorbar
plt.grid(True)
plt.show()
```

```python
# Apply Kernel PCA with RBF kernel
kpca_X = KernelPCA(n_components=2, kernel='rbf', gamma=0.002)  # Reduce to 2 dimensions
prcomps_kX = kpca_X.fit_transform(X)

plt.figure(figsize=(6, 4))
plt.scatter(prcomps_kX[:, 0], prcomps_kX[:, 1], c=t, s=10, cmap='hot_r')
plt.xlabel('Kernel PCA Dimension 1', fontsize=12)
plt.ylabel('Kernel PCA Dimension 2', fontsize=12)
plt.title("2D Kernel PCA of Swiss Roll Dataset", fontsize=14)
plt.colorbar(label='t')
plt.grid(True)
plt.show()
```
